### 1. Logica de Programação e Pseudocódigo

**O que é lógica de programação?**

Antes de escrever código em uma linguagem de programação, é fundamental entender a lógica por trás da solução de um problema. Um algorítimo é uma sequência de passos finítos e bem definidos para resolver um problema. O pseudocódigo e uma forma de representar de maneira informal e próxima da linguagem humana, antes de traduzí-los para Python.

**O que é Pseudocódigo**

Pseudocódigo é uma forma de descrer algorítmos usando uma forma de linguagem intermediária entre o português (ou outro idioma natural) e uma linguagem de programação real (como Python, Java, C++).

Ele não segue regras rígidas de sintaxe, mas utiliza estruturas comuns de programação (como se, então, senão, enquanto, para cada, função etc.), escritas de maneiras simples e legíveis para humanos.

A ideia é focar na lógica do algorítmo, sem se preocupar com os detalhes técnicos da linguágem que será usada depois.

Vejamos um exemplo de pseudocódigo:

Início

    LEIA nota1

    LEIA nota2

    CALCULE media = (nota1 + nota2) / 2

    ESCREVA "A média do aluno é:", media

    SE media >= 7.0 ENTÃO

        ESCREVA "Aluno Aprovado!"

        SENÃO

        ESCREVA "Aluno Reprovado."

    FIM

Tradução em Python:

```python
# Traduzindo o Pseudocódigo para Python

nota1 = 7.5
nota2 = 8.0

media = (nota1 + nota2) / 2

print(f"A média do aluno é: {media}")

if media >= 7.0:
    print("Aluno Aprovado")

else:
    print("Aluno Reprovado.")
```

### 2. Variáveis: Declaração, Atribuição e Regras de Nomenclatura

Uma variável é um espaço na memória do computador destinado a armazenar dados. Em Python, a declaração e a atribuição de um valor a uma variável são feitas simultaneamente. Regras de Nomenclaturas:

 - Nomes de variáveis devem começar com uma letra ou underscore(_).
 - Não podem começar com número.
 - Podem conter apenas caracteres alfanuméricos e underscores (A-z, 0-9 e _).
 - São "case-sensitive" (idade é diferente de Idade).

```python
# Declaração e Atribuição de Váriáveis
nome_completo = "Leonardo da Silva Neves" # String (Texto)
idade = 37 # Integer (inteiro)
altura = 1.87 # Float (ponto flutuante)
eh_estudante = True # Boolean (booleano)
```

```python
print(f"Nome: {nome_completo}")
print(f"Idade: {idade}")
print(f"altura: {altura}")
print(f"É estudante?: {eh_estudante}")
```

```python
# Exemplo de nomes de variáveis inválidas
# 1nome = "Inválido"
# nome-completo = "Inválido"
```

```python
x = 10 # tipo int
y1 = 10 # tipo int
y2 = "10" # tipo Str
```

### 3. Escopo de Variáveis

O escopo de uma variável define onde ela pode ser acessada no código.

 - Variáveis Globais: Declaradas fora de qualquer função. Podem ser acessadas de qualquer lugar do código.
 - Variáveis Locais: Declaradas dentro de uma função. Só podem ser acessadas dentro daquela função.

```python
# Variável Global
saudacao = "Olá mundo!"
nome = "Leonardo"

# Função
def funcao_exemplo():

    #Variável local
    nome = "Anastácia"
    print(f"Dentro da função: {nome}")
    print(f"\nAcessando a variável global de dentro da função: {saudacao}")

funcao_exemplo()

print(f"\nFora da função: {saudacao}")
print(f"\nFora da função: {nome}")
```

```python
# Função
def funcao_exemplo():

    #Variável local
    nome_funcao = "Anastácia"
    print(f"Dentro da função: {nome}")
    print(f"\nAcessando a variável global de dentro da função: {saudacao}")

funcao_exemplo()

# Tentar acessar a variável declarada dentro da função resultará em um erro!
# print(f"\nFora da função: {nome_funcao}")
```

### 5. Operadores Aritméticos, de Comparação e Lógicos

#### 5.1 Operadores Aritméticos

Usados para realizar operações matemáticas.

```python
# Definição de variáveis
a = 10
b = 3
```

```python
# Usando operadores aritiméticos
soma = a + b # Adição
subtracao = a - b # Subtração
multiplicacao = a * b # Multiplicação
divisao = a / b # Divisão com resultado sempre float
divisao_inteira = a // b # Divisão inteira (descarta a parte decimal)
modulo = a % b # Módulo (resto da divisão)
potencia = a ** b # Potenciação
```

```python
print(f"{a} + {b} = {soma}")
print(f"{a} - {b} = {subtracao}")
print(f"{a} * {b} = {multiplicacao}")
print(f"{a} / {b} = {divisao:.2f}")
print(f"{a} // {b} = {divisao_inteira}")
print(f"{a} % {b} = {modulo}")
print(f"{a} ** {b} = {potencia}")
```

```python
# As regras da matemática se aplicam aqui normalmente

# Variáveis

a = 10
b = 0

# Tentativa de divisão por zero
# a/b
```

```python
# Cuidado. Isso não pode!
# 8 + 's'
```

```python
# Mas isso é permitido! (não é soma, é concatenação de strings)
'8' + 's'
```

#### 5.2. Operadores de Comparação

Usandos para comparar valores. O resultado é sempre um Boolean (True ou False).

```python
# Definição de variáveis
x = 5
y = 10
```

```python
# Operador "maior que"
x > y
```

```python
# Operador "Menor que"
x < y
```

```python
# Operador "igual a"
x == y
```

```python
# Operador "Diferennte de"
x != y
```

```python
# Operador "Maior ou igual a"
x >= y
```

```python
# Operador "Menor ou igual a"
x <= y
```

```python
print(f"{x} > {y} ? {x>y}") # Maior que
print(f"{x} < {y} ? {x<y}") # Menor que
print(f"{x} == {y} ? {x==y}") # Igual a
print(f"{x} != {y} ? {x!=y}") # Diferente a
print(f"{x} >= {y} ? {x>=y}") # Maior ou Iqual a
print(f"{x} <= {y} ? {x<=y}") # Menor ou Igual a
```

#### 5.3. Operadores Lógicos

Usados para combinar expressões booleanas.

```python
# Definição de variáveis
tem_dinheiro = True
tem_tempo = False
```

```python
# Operador AND (e): Ambos precisam ser verdadeiros
print(f"O cliente pode viajar? {tem_dinheiro and tem_tempo}")
```

```python
# Operador OR (ou): Pelo menos um precisa ser verdadeiro
print(f"O cliente pode viajar? {tem_dinheiro or tem_tempo}")
```

```python
# Operador NOT (não): Inverte o valor booleano
print(f" O cliente pode viajar? {tem_dinheiro and not tem_tempo}")
```

#### 10.Estrutura de Dados - Conjuntos (Sets)

Conjuntos são coleções não ordenadas de itens únicos e mutáveis. Eles são úteis para remover duplicatas e realizar operaçẽos matemáticas de conjuntos(união, interseção).

```python
# Criando um conjunto (note que os itens duplicados são removidos)
numeros = {1, 2, 3, 4, 2, 3, 5}
print(f"Conjunto de números (sum duplicatas): {numeros}")
```

```python
type(numeros)
```

```python
# Adicionando um item
numeros.add(2)
print(f"Após adicionar o número 6: {numeros}")
```

```python
# Remover um item
numeros.remove(2)
print(f"Após remover o 2: {numeros}")
```

```python
# Operação de conjutos
conjunto_a = {1, 2, 3, 4}
conjunto_b = {3, 4, 5, 6}
```

```python
# União (todos os elementos de ambos os conjuntos)
uniao = conjunto_a.union(conjunto_b)
print(f"União de A e B: {uniao}")
```

```python
# Interseção (elementos que estão em ambos os conjutos)
intersecao = conjunto_a.intersection(conjunto_b)
print(f"Interseção de A e B: {intersecao}")
```

#### 11. Conversão Entre Tipos de Dados (Type Casting)

É a conversão de um tipo de dado pora outro.

```python
# Conversão de string para o tipo integer
numero_em_texto = "123"
numero_inteiro = int(numero_em_texto)
print(f"String '{numero_em_texto}' para Integer: {numero_inteiro}, Tipo {type(numero_inteiro)}")
```

```python
# Convertendo de string para número float
numero_decimal_em_texto = "45.67"
numero_float = float(numero_decimal_em_texto)
print(f"String: '{numero_decimal_em_texto}' para Float: {numero_float}, Tipo: {type(numero_float)}")
```

```python
# Convertendo de número para string
idade = 25
idade_texto = str(idade)
print(f"Inteiro: {idade} para String: '{idade_texto}', Tipo: {type(idade_texto)}")
```

```python
# Convertendo uma estrutura de dados
lista_com_duplicatas = [1, 2, 2, 3, 4, 4, 4, 5]
conjunto_unico = set(lista_com_duplicatas)
lista_sem_duplicatas = list(conjunto_unico)
```

```python
print(f"\nLista original: {lista_com_duplicatas}")
print(f"\nConvertida para conjunto (duplicatas removida): {conjunto_unico}")
print(f"\nConvertida de volta para Lista: {lista_sem_duplicatas}\n")
```

##### 12.2. Entrada de Dados input()

A função input() sempre retorna uma string. Por isso, é comum precisar fazer o type casting.

```python
# Pedido o nome do usuário (string)
nome_usuario = input("Qual é o seu nome? ")

# Pedindo a idade do usuário (precisa converter para int)
idade_usuario_str = input("Qual é a sua idade? ")
idade_usuario_int = int(idade_usuario_str)

from datetime import date

# Pega o ano corrente na data definida no sistema operacional da sua máquina.
ano_atual = date.today().year

# Preenchendo os dados
ano_nascimento = ano_atual - idade_usuario_int

# Exibindo o resultado
print(f"\nOlá, {nome_usuario}! Bem-vindo(a).")
print(f"Você tem {idade_usuario_int} anos e nasceu aproximadamente em {ano_nascimento}.")
```

```python
altura_str = input("Digite sua altura em metro: Ex( 1.34): ")
peso_str = input("Digite seu peso em kg (ex: 70,5): ")

altura = float(altura_str)
peso = float(peso_str)

imc = peso / (altura ** 2)

print(f"IMC calculado: {imc}")
```
