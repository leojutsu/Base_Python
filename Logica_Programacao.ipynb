### Fundamentos de Linguagem Python para Construção de Jogos

Com base nos conceitos estudados até aqui, vamos criar um jogo clássico e simples: Pedra, Papel e Tesaura, para dois jogadores.

Este estudo de cado é ideal para iniciantes, pois utiliza apenas os fundamento ensinados: variáveis, entrada e saída, manipulação de strings para normalizar as entradas dos jogadores, e estruturas condicionais (if, elif, else) com operadores lógicos para determinar o vencedor.

Pedra, Papel, Tesoura é um jogo de mão simples e popular, geralmente disputado entre duas pessoas. O obejetivo é derrotar o oponente escolhendo um de três objetos.

As regras de vitória são as seguintes:

- Pedra vence tesoura (quebrando-a)
- Tesoura vence papael (cortando-o)
- Papel vence pedra (cobrindo-o)

Se ambos os jogadores escolherem o mesmo objeto, a rodada resulta em empate.

Por ser um jogo rápido e baseado principalmente na sorte, é frequentemente usado para tomar decisões triviais ou para decidir quem começa em outra atividade.

**2. Script Python**

```python
# 1. Apresentação das Regras

# Usando a função print() para exibir mensagens na tela (Saída padrão).
print(40 * "-")
print("Jogo Pedra, Papel, Tesoura (2 Jogadores)")
print(40 * "-")
print("Bem-vindo! Cada jogador deve escolher uma das opções.")

# Usando tuplas para armazenar as opções válidas.
# Tuplas são boas porque as opções de jogadas não mudam (são imutáveis)
opcoes_validas = ("pedra", "papel", "tesoura")
print(f"Opções válidas: {opcoes_validas}")
print("-"*40)

# 2. Coleta dos dados de enntrada

# Usando a função input() para receber o que o usuário digita (Entrada padrão)
# A função input() sempre retorna um string.

jogada_jogador1_inicial = input("Jogador 1, digite sua jogada: ")
jogada_jogador2_inicial = input("Jogador 2, digite sua jogada: ")

# 3. Tratamento dos Dados de entrada

# Manipulação de strings usasndo o .lower() para converter para minusculo
# Manipulação de strings usando o .strip() para remover os espaços

jogada_jogador1 = jogada_jogador1_inicial.lower().strip()
jogada_jogador2 = jogada_jogador2_inicial.lower().strip()

print(30 * "-")
print(f"jogador 1 escolheu: {jogada_jogador1}")
print(f"jogador 2 escolher: {jogada_jogador2}")
print(30 * "-")

# 4. Lógica do jogo e resultado

# Primeiro verificar se as jogadas são válidas

if jogada_jogador1 not in opcoes_validas or jogada_jogador2 not in opcoes_validas:
    print("Uma das jogadas são inválidas! Por favor, use apenas 'Pedra', 'Papel' ou 'Tesoura', ")

# Lógica principal do jogo

elif jogada_jogador1 == jogada_jogador2:
    print("Resultado: É um empate!")
# Caso 2: Jogador 1 vence
# Agrupando todas as codições de vitória do jogador 1 com o operador 1.
elif (jogada_jogador1 == "pedra" and jogada_jogador2 == "tesoura") or \
     (jogada_jogador1 == "tesoura" and jogada_jogador2 == "papel") or \
     (jogada_jogador1 == "papel" and jogada_jogador2 == "pedra"):
    print("Resultado: Jogador 1 venceu! Parabéns!")

# Caso 3: Se não empatou e jogador 1 não venceu
else:
    print("Resultado: Jogador 2 venceu! Parabéns!")

print( "\n--- Fim de Jogo ---")
```
