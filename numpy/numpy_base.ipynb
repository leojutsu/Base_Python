**1. Começando comparando a performance do Numpy com Python puro**

```python
# )mports
import math
import time
```

```python
precos_np = np.random.rand(10_000_000)
type(precos_np)
```

```python
precos_list = list(precos_np)
type(precos_list)
```

```python
# Operação com Numpy
t0 = time.time()
desc = precos_np * 0.90
final = desc + 5
raiz = np.sqrt(precos_np)
print("NumPy:", time.time() - t0, "segundos")
```

```python
# Mesma operação em Python puro
t0 = time.time()
desc = [p * 0.9 for p in precos_list]
final = [p + 5 for p in desc]
raiz = [math.sqrt(p) for p in precos_list]
print("Python puro", time.time() - t0, "segundos")
```

Para dez milhoẽs de valores, o Numpy normalmente é mais rápido. Mas para poucos valores, o uso de loop com Python puro pode sair na frente porque o overhead do Numpy pesa mais que o ganho para poucos registros.

**Resumindo**

- Poucos elemendos -> Loops Python podem ser mais rápidos (overhead do Numpy é maior).
- Muitos elementos -> Numpy é muito mais rápido e escalável.

É pela sua velocidade, que Numpy é amplamente usado em projetos de Ciência de Dados, Machine Learning e IA.
