**1. Tomadas de Decisões com Estruturas Condicionais**

As estruturas condicionais (if, elif, else) permite que o programa execute diferentes blocos de código com base em certas condições.

```python
# Definindo a variável

nota = 8.5

# Agora checamos o valor da variável para tomar a decisão
if nota >= 7.0:
    print("Aprovado!")
else:
    print("Reprovado.")
```

```python
# Definindo a variável
idade = 17

# Agora checamos o valor da variável e tomamos uma decisão
if idade < 18:
    print("Você é menor de idade.")
elif idade >= 18 and idade < 65:
    print("Você é adulto.")
else:
    print("Você é um idoso.")
```

**2. Estruturas de Repetição**

As estruturas de repetição (for e while) são usadas para executar um bloco de código várias vezes.

**Loop for**

O loop for é usado para interar sobre uma sequência  (como uma lista, tupla, dicionário, conjunto ou string).

```python
# Definindo uma lista
frutas = ["maça", "banana", "cereja"]

# Imprime a mensagem
print("Frutas disponíveis:")

# Loop pela lista
for fruta in frutas:
    print(f"- {fruta}")
```

```python
# Criando uma tupla
cores = ("vermelho", "verde", "azul")

# Loop for percorrendo a tupla
for cor in cores:
    print(f"- {cor}")
```

```python
# Criando um dicionário com o número de cursos em formação da DSA
formacoes = {"Formação Cientista de Dados": 6, "Formação Analista de Dados": 4, "Formação Engenharia de Dados": 5}

# Loop for percorrendo chaves e valores
for chave, valor in formacoes.items():
    print(chave, ":", valor)
```

```python
# Exemplo com a função range()
print("\nContagens até 5:")
for numero in range(6): # Gera números de 0 a 5
    print(numero)
```
